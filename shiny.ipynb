```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('/content/download.jpg')
flipped_horizontally=cv2.flip(image,1)#horizontal
flipped_vertically=cv2.flip(image,0)#vertical shift
flipped_both=cv2.flip(image,-1)#vertical mirror image

plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("original")
plt.subplot(2,3,2)
tx,ty=10,-10#shift in x and y direction
translation_matrix=np.float32([[1,0,tx],[0,1,ty]])#apply translation
translated=cv2.warpAffine(image,translation_matrix,(w,h))
plt.imshow(cv2.cvtColor(translated,cv2.COLOR_BGR2RGB))
plt.title("translated image")


scale_x,scale_y=5,10
resized=cv2.resize(image,None,fx=scale_x,fy=scale_y,interpolation=cv2.INTER_LINEAR)


plt.subplot(2,3,3)#(row,column,2ndtimage)
plt.imshow(cv2.cvtColor(resized,cv2.COLOR_BGR2RGB))
plt.title("scaled image")
plt.subplot(2,3,4)
(h,w)=image.shape[:2]
center=(20,60)
angle=360
scale=1.0
rotation_matrix=cv2.getRotationMatrix2D(center,angle,scale)
rotated=cv2.warpAffine(image,rotation_matrix,(w,h))
plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.title("Rotated Image(360°)")
plt.subplot(2,3,5)
shear_matrix=np.float32([[0,-1,0],[2,-1,0]])
sheared_image=cv2.warpAffine(image,shear_matrix,(w,h))
plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.title("sheared image")
plt.show()
```

```python
from google.colab.patches import cv2_imshow
import cv2
from google.colab import files
from IPython.display import clear_output
import time
video_path="/content/as.mp4.mp4"
cap=cv2.VideoCapture(video_path)
if not cap.isOpened():
  print("Error opening video file")
fps=cap.get(cv2.CAP_PROP_FPS)
delay=1/fps
while cap.isOpened():
  ret, frame=cap.read()
  if not ret:
    print("end of video")
    break
  clear_output(wait=True)
  cv2_imshow(frame)
  time.sleep(delay)
cap.release()
```
